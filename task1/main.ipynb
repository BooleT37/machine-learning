```python
import pandas
from sklearn import svm
import numpy as np
from sklearn.model_selection import cross_val_score
```

# Load training data

```python
df = pandas.read_csv('data/task.csv', ';')
df = df.drop('dem_cand', axis=1).drop('rep_cand', axis=1)
df_train = df[(df.pidAggr == 1) | (df.pidAggr == 2)]
X_train = df_train.drop(['pidAggr'], axis=1).astype(int)
X_train.head()
```

```python
y_train = df_train[['pidAggr']].astype(int).as_matrix().ravel()
y_train
```

# Cross-validate data

```python
clf = svm.SVC(kernel='linear')
clf.fit(X_train, y_train)
clf = svm.SVC(kernel='linear', C=1)
scores = cross_val_score(clf, X_train, y_train, cv=5)
scores
```

# Predict pidAggr

```python
clf.fit(X_train, y_train)
X = df[df['pidAggr'].isna()].drop('pidAggr', axis=1)
prediction = np.stack((X['id'], clf.predict(X)), axis=1)
prediction_str = '\n'.join(';'.join(str(cell) for cell in row) for row in prediction)
print(prediction_str)
```

# Save results to file

```python
file = open('out.csv', 'w')
file.write(prediction_str)
file.close()
```
